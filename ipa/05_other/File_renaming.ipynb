```python
import glob
import os
import shutil

import numpy as np

%load_ext autoreload
%autoreload 2
```

```python
# Import images

img_merged_dir = '/Volumes/gchao/bamfaile/Analysis/ACTB-KI/00_Batch20241011_iNs/D10/Unperturbed/50ms/Denoised_merged'
img_merged_path = os.path.join(img_merged_dir,'*.tif')
img_merged_files = np.sort(glob.glob(img_merged_path))

mask_dir = '/Volumes/gchao/bamfaile/Analysis/ACTB-KI/00_Batch20241011_iNs/D10/Unperturbed/ROIs/Cytoplasm'
mask_dir = os.path.join(mask_dir,'*.tif')
mask_files = np.sort(glob.glob(mask_dir))
```

```python
print(len(img_merged_files))
print(len(mask_files))
```

```python
mask_files
```

```python
# Function to extract the integer before the first underscore
def extract_integer(file_name):
    return int(file_name.split('_')[0])

# Iterate over both file lists
for img_file, mask_file in zip(img_merged_files, mask_files):
    # Extract the integers from the file names
    img_int = extract_integer(os.path.basename(img_file))
    mask_int = extract_integer(os.path.basename(mask_file))
    
    # Ensure the integers match
    if img_int == mask_int:
        # Extract the base name from the img_merged_file without the extension
        base_name = os.path.splitext(os.path.basename(img_file))[0]
        # Create the new name for the mask file
        new_name = f"{base_name}_ROIs.tif"
        # Construct the full path for the new name
        new_path = os.path.join(os.path.dirname(mask_file), new_name)
        
        # Rename the mask file
        shutil.move(mask_file, new_path)
        print(f"Renamed: {mask_file} -> {new_path}")
    else:
        # Log a mismatch error
        print(f"Mismatch found: {img_file} and {mask_file}. Skipping rename.")
```
