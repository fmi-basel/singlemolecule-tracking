```python
# Loads mean intensites of each ROI per image

D14_allROIs_meanint_paths = ['/Volumes/gchao/bamfaile/Analysis/ACTB-KI/Batch20220930/D14/Slide1/ROIs/Meanint_ROIs_pos14.csv',
                             '/Volumes/gchao/bamfaile/Analysis/ACTB-KI/Batch20220930/D14/Slide1/ROIs/Meanint_ROIs_pos19.csv']

D14_allROIs_meanint = []

for path in D14_allROIs_meanint_paths:
    df = load_meanint_ROIs(path)
    D14_allROIs_meanint.append(df)
    
D14_allROIs_meanint[1].head()
```

```python
# Removes tracks that are <= 3 from tracks file

D14_allROIs_tracks_filtered = []

for df in D14_allROIs_tracks:
    df = remove_shorttracks_3(df)
    D14_allROIs_tracks_filtered.append(df)

#pd.set_option('display.max_rows', None) # Displays all rows of df    
D14_allROIs_tracks_filtered[5].head()
```

```python
# Creates trackIDs object 

D14_allROIs_trackIDs = []

for df in D14_allROIs_tracks_filtered:
    ID = df["TRACK_ID"].unique()
    D14_allROIs_trackIDs.append(ID)

D14_allROIs_trackIDs[1]
```

```python
# Removes spots from spots file from tracks <= 3 in length

D14_allROIs_spots_filtered = []


for df, i in zip(D14_allROIs_spots, D14_allROIs_trackIDs):
    
    df = remove_shorttracks_spotsfile(df, i)
    D14_allROIs_spots_filtered.append(df)


#pd.set_option('display.max_rows', None) # Displays all rows of df  
D14_allROIs_spots_filtered[4].head(10)
```

```python
# Splits tables into seperate dfs by TRACK_ID

D14_pos14_ROI1_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[0])]
D14_pos14_ROI2_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[1])]
D14_pos14_ROI3_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[2])]
D14_pos19_ROI1_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[3])]
D14_pos19_ROI2_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[4])]
D14_pos19_ROI3_singletracks = [j for i,j in split_table_by_track(D14_allROIs_spots_filtered[5])]

D14_allROIs_singletracks = [D14_pos14_ROI1_singletracks, 
                            D14_pos14_ROI2_singletracks,
                            D14_pos14_ROI3_singletracks, 
                            D14_pos19_ROI1_singletracks,
                            D14_pos19_ROI2_singletracks,
                            D14_pos19_ROI3_singletracks]

D14_allROIs_singletracks[5][5].head(10)
```

```python
# Sorts the seperated dfs by FRAME and extracts the first 10

D14_pos14_ROI1_tracks10 = []
D14_pos14_ROI2_tracks10 = []
D14_pos14_ROI3_tracks10 = []
D14_pos19_ROI1_tracks10 = []
D14_pos19_ROI2_tracks10 = []
D14_pos19_ROI3_tracks10 = []

D14_allROIs_tracks10 = [D14_pos14_ROI1_tracks10, 
                        D14_pos14_ROI2_tracks10, 
                        D14_pos14_ROI3_tracks10, 
                        D14_pos19_ROI1_tracks10,
                        D14_pos19_ROI2_tracks10,
                        D14_pos19_ROI3_tracks10]

for i in D14_allROIs_singletracks:
    index = D14_allROIs_singletracks.index(i)
    for j in i:
        j = first_10_frames(j)
        D14_allROIs_tracks10[index].append(j)

D14_allROIs_tracks10[0][2].head(11)
```

```python
# Concatenates all filtered seperate track dfs into one df per ROI

D14_allROIs_alltracks10 = []

for i in D14_allROIs_tracks10:
    for df in i:
        df = pd.concat(i)
    D14_allROIs_alltracks10.append(df)
D14_allROIs_alltracks10[4].head()
```

```python
# Extracts the columns "TRACK_ID" and "MEAN_INTENSITY_CH2" and calculates the mean of each track

D14_allROIs_meanint = []

for df in D14_allROIs_alltracks10:
    df = meanint_tracks_ch2(df)
    D14_allROIs_meanint.append(df)
    
D14_allROIs_meanint[5].head()
```
