```python
# Removes tracks <= 3 from tracks file

D14_pos14_allROIs_tracks_filt = []


for df in D14_pos14_allROIs_tracks_clean:
    
    df["NUMBER_SPOTS"] = pd.to_numeric(df["NUMBER_SPOTS"])
    df['NUMBER_SPOTS'] = df['NUMBER_SPOTS'].astype(int)
    
    df = df[df["NUMBER_SPOTS"] >= 3].reset_index(drop=True)
    D14_pos14_allROIs_tracks_filt.append(df)
    
D14_pos19_allROIs_tracks_filt = []

for df in D14_pos19_allROIs_tracks_clean:
    
    df["NUMBER_SPOTS"] = pd.to_numeric(df["NUMBER_SPOTS"])
    df['NUMBER_SPOTS'] = df['NUMBER_SPOTS'].astype(int)
    
    df = df[df["NUMBER_SPOTS"] >= 3].reset_index(drop=True)
    D14_pos19_allROIs_tracks_filt.append(df)
    
D14_pos19_allROIs_tracks_filt[0].head()
```

```python
# Creates trackIDs object 

D14_pos14_allROIs_trackIDs = []

for df in D14_pos14_allROIs_tracks_filt:
    ID = df["TRACK_ID"].unique()
    D14_pos14_allROIs_trackIDs.append(ID)
    
D14_pos19_allROIs_trackIDs = []

for df in D14_pos19_allROIs_tracks_filt:
    ID = df["TRACK_ID"].unique()
    D14_pos19_allROIs_trackIDs.append(ID)
    

D14_pos14_allROIs_trackIDs[0]
```

```python
# Filters tracks in spots file by trackID object to remove all tracks <=3 from spots file

D14_pos14_ROI1_filt = D14_pos14_allROIs_clean[0].loc[D14_pos14_allROIs_clean[0]['TRACK_ID'].isin(D14_pos14_allROIs_trackIDs[0])]
D14_pos14_ROI2_filt = D14_pos14_allROIs_clean[1].loc[D14_pos14_allROIs_clean[1]['TRACK_ID'].isin(D14_pos14_allROIs_trackIDs[1])]
D14_pos14_ROI3_filt = D14_pos14_allROIs_clean[2].loc[D14_pos14_allROIs_clean[2]['TRACK_ID'].isin(D14_pos14_allROIs_trackIDs[2])]

D14_pos19_ROI1_filt = D14_pos19_allROIs_clean[0].loc[D14_pos19_allROIs_clean[0]['TRACK_ID'].isin(D14_pos19_allROIs_trackIDs[0])]
D14_pos19_ROI2_filt = D14_pos19_allROIs_clean[1].loc[D14_pos19_allROIs_clean[1]['TRACK_ID'].isin(D14_pos19_allROIs_trackIDs[1])]
D14_pos19_ROI3_filt = D14_pos19_allROIs_clean[2].loc[D14_pos19_allROIs_clean[2]['TRACK_ID'].isin(D14_pos19_allROIs_trackIDs[2])]

D14_pos14_allROIs_filtered = [D14_pos14_ROI1_filt,D14_pos14_ROI2_filt,D14_pos14_ROI3_filt]
D14_pos19_allROIs_filtered = [D14_pos19_ROI1_filt,D14_pos19_ROI2_filt,D14_pos19_ROI3_filt]

D14_allROIs_filtered = [D14_pos14_ROI1_filt,
                        D14_pos14_ROI2_filt,
                        D14_pos14_ROI3_filt,
                        D14_pos19_ROI1_filt,
                        D14_pos19_ROI2_filt,
                        D14_pos19_ROI3_filt]

D14_allROIs_filtered[0].head()
```

```python
# Splits each df into sperate dfs by TRACK_ID, sorts the rows by FRAME and filters first 10 rows

D14_pos14_ROI1_singletracks = [j for i,j in D14_pos14_ROI1_filt.groupby("TRACK_ID", sort = False, as_index = False)] #Why does this line work?
D14_pos14_ROI2_singletracks = [j for i,j in D14_pos14_ROI2_filt.groupby("TRACK_ID", sort = False, as_index = False)] 
D14_pos14_ROI3_singletracks = [j for i,j in D14_pos14_ROI3_filt.groupby("TRACK_ID", sort = False, as_index = False)]

D14_pos19_ROI1_singletracks = [j for i,j in D14_pos19_ROI1_filt.groupby("TRACK_ID", sort = False, as_index = False)]
D14_pos19_ROI2_singletracks = [j for i,j in D14_pos19_ROI2_filt.groupby("TRACK_ID", sort = False, as_index = False)]
D14_pos19_ROI3_singletracks = [j for i,j in D14_pos19_ROI3_filt.groupby("TRACK_ID", sort = False, as_index = False)]

D14_allROIs_singletracks = [D14_pos14_ROI1_singletracks, 
                            D14_pos14_ROI2_singletracks,
                            D14_pos14_ROI3_singletracks, 
                            D14_pos19_ROI1_singletracks,
                            D14_pos19_ROI2_singletracks,
                            D14_pos19_ROI3_singletracks]

D14_pos14_ROI1_tracks10 = []
D14_pos14_ROI2_tracks10 = []
D14_pos14_ROI3_tracks10 = []
D14_pos19_ROI1_tracks10 = []
D14_pos19_ROI2_tracks10 = []
D14_pos19_ROI3_tracks10 = []

D14_allROIs_tracks10 = [D14_pos14_ROI1_tracks10, 
                        D14_pos14_ROI2_tracks10, 
                        D14_pos14_ROI3_tracks10, 
                        D14_pos19_ROI1_tracks10,
                        D14_pos19_ROI2_tracks10,
                        D14_pos19_ROI3_tracks10]


for i in D14_allROIs_singletracks:
    index = D14_allROIs_singletracks.index(i)
    for j in i:
        j['FRAME'] = j['FRAME'].astype(int) # Is probably necessary to have proper ordering of tracks by frame
        j = j.sort_values(by="FRAME").iloc[0:10]
        D14_allROIs_tracks10[index].append(j)
    
    
D14_allROIs_tracks10[0][25].head(10)
```

```python
# Concatenates all filtered seperate track dfs into one df per ROI

D14_allROIs_alltracks10 = []

for i in D14_allROIs_tracks10:
    for df in i:
        df = pd.concat(i)
    D14_allROIs_alltracks10.append(df)
D14_allROIs_alltracks10[4].head(25)
```

```python
# Extracts columns "Track ID" and "Mean intensity" and calculates the mean intensity of each track

D14_allROIs_track_int_mean = []

for df in D14_allROIs_alltracks10:
    
    """
    .astypefloat converts values of 'MEAN_INTENSITY_CH2' to type 'float', so that 'groupby' operation in next line can be perfomed
    .iloc selects the columns TRACK_ID and MEAN_INTENSITY_CH2; the ':' in the square brackets signifies that the rows are left untouched
    .mean calculates mean intensities based on .groupby 'TRACK_ID'
    .round rounds the values to 2 decimals after calculating the mean
    
    """
        
    df['MEAN_INTENSITY_CH2'] = df['MEAN_INTENSITY_CH2'].astype(float)
    df = df.iloc[:, np.r_[2,21]].groupby("TRACK_ID", sort = False, as_index = False).mean().round(2)
    D14_allROIs_track_int_mean.append(df)

D14_allROIs_track_int_mean[0].head()   
```

```python
# Concatenates dfs of D14_allROIs_track_int_mean into one df and into two dfs per ROI

D14_track_int_mean = pd.concat(D14_allROIs_track_int_mean)
D14_pos14_track_int_mean = pd.concat(D14_allROIs_track_int_mean[0:3])
D14_pos19_track_int_mean = pd.concat(D14_allROIs_track_int_mean[3:6])

D14_pos14_track_int_mean.head()
```
